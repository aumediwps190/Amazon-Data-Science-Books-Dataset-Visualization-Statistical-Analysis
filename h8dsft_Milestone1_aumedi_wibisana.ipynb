# Data Cleaning

In this section, I will remove the '[ ]' on the author names. The NaN in all columns that are meant to be numbers will be replaced with 0. The '%' will be removed and all the numbers in `star5` to `star1` will be converted to float and divided by 100.

```python
import numpy as np
```

```python
dfess.info()
```

```python

# Removing the '[' and ']' in author

dfess['author'] = dfess['author'].replace({'\[': ''}, regex=True)
dfess['author'] = dfess['author'].replace({'\]': ''}, regex=True)



# Turning all the NaNs into zero

dfess['price'] = dfess['price'].replace({np.nan: float(0)}, regex=True)
dfess['pages'] = dfess['pages'].replace({np.nan: int(0)}, regex=True)
dfess['avg_reviews'] = dfess['avg_reviews'].replace({np.nan: float(0)}, regex=True)
dfess['n_reviews'] = dfess['n_reviews'].replace({np.nan: int(0)}, regex=True)
dfess['star5'] = dfess['star5'].replace({np.nan: float(0)}, regex=True)
dfess['star4'] = dfess['star4'].replace({np.nan: float(0)}, regex=True)
dfess['star3'] = dfess['star3'].replace({np.nan: float(0)}, regex=True)
dfess['star2'] = dfess['star2'].replace({np.nan: float(0)}, regex=True)
dfess['star1'] = dfess['star1'].replace({np.nan: float(0)}, regex=True)


# Dealing with n_reviews

dfess['n_reviews'] = dfess['n_reviews'].replace({'\,': ''}, regex=True)
dfess = dfess.astype({'n_reviews': float})


# Dealing with '%' in star5 to star1 columns as well as turning them into floats
# before dividing them by 100

dfess['star5'] = dfess['star5'].replace({'%': ''}, regex=True)
dfess['star4'] = dfess['star4'].replace({'%': ''}, regex=True)
dfess['star3'] = dfess['star3'].replace({'%': ''}, regex=True)
dfess['star2'] = dfess['star2'].replace({'%': ''}, regex=True)
dfess['star1'] = dfess['star1'].replace({'%': ''}, regex=True)

dfess = dfess.astype({'star5': float})
dfess = dfess.astype({'star4': float})
dfess = dfess.astype({'star3': float})
dfess = dfess.astype({'star2': float})
dfess = dfess.astype({'star1': float})

dfess['star5'] = dfess['star5'] / 100
dfess['star4'] = dfess['star4'] / 100
dfess['star3'] = dfess['star3'] / 100
dfess['star2'] = dfess['star2'] / 100
dfess['star1'] = dfess['star1'] / 100

dfess
```

Now I want to figure out the price range that might works best for my service. I will find the average price as well as the standard deviation. Then I will find the skewness of the price distribution. This will help in identifying outliers lurking in the data. I will then reevaluate based on updated dataset with no outliers.

```python
# Mean, STD, and skewness of price

print("Mean of price:", db['price'].mean())
print("STD of price:", db['price'].std())
print("Skewness of price:", db['price'].skew())
```

The average prices for each titles is 42.22 USD with the standard deviation of 31.68. Since the skewness is greater than |+-0.5|, I decide to utilize Tukey's rule to identify the outliers. The first code block will present the number while the second one will visualize.

```python
# Identifying outliers

# Tukey's Rule used to determine the upper and lower bound
q1 = db['price'].quantile(0.25)
q3 = db['price'].quantile(0.75)
iqr = q3-q1

up_bound = q3 + 1.5*iqr
low_bound = q1 - 1.5*iqr

print(f'Upper bound: {up_bound:.2f}' )
print(f'Lower bound: {low_bound:.2f}')

outliersPrice = db[(db['price'] < low_bound) | (db['price'] > up_bound)]
print("Numbers of outliers in price:", len(outliersPrice))
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(10,4))

db['price'].plot(kind='hist',bins=20,ax=ax[0])
db['price'].plot(kind='box',ax=ax[1])
```

There are 40 outliers in the price series and all of them are above the upper bound. I will now filter out the outlier from the data.

```python
# Filtering the outliers

dbNoOutlier = db[(db['price'] >= low_bound) & (db['price'] <= up_bound)]
dbNoOutlier
```

The number of entry (rows) is consistent since 621 - 40 = 581. Let's see what's the average price is without the outliers. The code below will print them.

```python
# Finding mean, std, and range for outlier-filtered dataset

print("Average price (No Outliers):", dbNoOutlier['price'].mean())
print("STD price (No Outliers):", dbNoOutlier['price'].std())
```

I can set my service (including the fee) at about 36 USD. But I should be willing to adjust this price to certain values depending on the books that the customers are willing to acquire.

```python
# Finding possible maximum and minimum price for my service

print("Possible maximum:", dbNoOutlier['price'].max())
print("Possible minimum:", dbNoOutlier['price'].min())
```

I have to consider raising the price up to 90 USD or lowering it to 1 USD depending on the book that the customers are getting. But it is unlikely that customers will acquire a book worth of 1 USD. No books worth that low.

Let's move on with the analysis and answer the last question in the Problem Breakdown section. To find whether the price influence the buyers, that is n_reviews, I have to find a correlation between the two series. I have to use a Chi-Square test. The hypothesis are given below:\

**H0** : There *isn't* a correlation between `price` and `n_reviews`\
**H1** : There *is* a correlation between `price` and `n_reviews`

```python
from scipy import stats
```

```python
contTable = pd.crosstab(dbNoOutlier['price'], dbNoOutlier['price'])

res = stats.chi2_contingency(contTable)
print(res) # The p-value is 0
```

Since the p-value is zero (which is smaller than 0.05) **H0** is rejected. There *is* a correlation between the data-science books' prices and the amount of copies sold. As is shown by the previous graph, the top ten most expensive data-science books didn't exceed 170 sales. The price has to be reasonable enough for customers to buy the books. This means that the prices does have an influence on the buyers.
